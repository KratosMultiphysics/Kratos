the Stokes equations are

$\frac{\partial\rho u}{\partial t}-\nabla\cdot\mathbf{C}\nabla^{s}\mathbf{u}+\nabla p=f$

$\frac{\partial\rho}{\partial t}+\nabla\cdot\left(\rho\mathbf{\mathbf{u}}\right)=0$

we consider the following (multiscale) decomposition

$\mathbf{u}=\mathbf{u_{h}}+\mathbf{u}_{s}$

$p=p_{h}+p_{s}$

where $\mathbf{u_{\mathbf{h}}}$ belongs to the finite element space.
If we now apply the Galerkin method and test the initial equations
by $w$ and $q$, we obtain

$\left(\mathbf{w},f+\nabla\cdot\mathbf{C}\nabla^{s}\mathbf{u}-\nabla p\right)+\left(q,-\frac{\partial\rho}{\partial t}-\nabla\cdot\left(\rho\mathbf{\mathbf{u}}\right)\right)=0$

substituting the proposed decomposition
\begin{align*}
\left(\mathbf{w},f-\frac{\partial\rho(\mathbf{u_{h}+\mathbf{u}_{s}})}{\partial t}+\nabla\cdot\mathbf{C}\nabla^{s}\left(\mathbf{u}_{h}+\mathbf{u_{s}}\right)-\nabla(p_{h}+p_{s})\right) & +\\
+\left(q,-\frac{\partial\rho}{\partial t}-\nabla\cdot\left(\mathbf{\rho u}_{h}\mathbf{+\rho u}_{s}\right)\right) & =0
\end{align*}


Integrating by parts the convective term

\begin{align*}
\left(\mathbf{w},\nabla\cdot C\nabla^{s}\left(\mathbf{u}_{h}+\mathbf{u_{s}}\right)\right)=-\left(\mathbf{\nabla w},C\nabla^{s}\mathbf{u}_{h}\right)+\int_{\Gamma}\mathbf{w}\cdot C\nabla^{s}\mathbf{u}_{h}\mathbf{\cdot n}+ & \left(\mathbf{w},2\nabla\cdot\mu\nabla^{s}\mathbf{u_{s}}\right)
\end{align*}
Here the second term will only contribute to the external boundary 

\begin{align*}
\left(\mathbf{w},-\nabla(p_{h}+p_{s})\right) & =\left(\mathbf{\nabla\cdot w},p_{h}+p_{s}\right)-\int_{\Gamma}p_{h}\mathbf{w}\mathbf{\cdot n}
\end{align*}


where we dropped the boundary term on $p_{s}$

Developing the continuity equation term and, again, using integration
by parts within each element and ommitting null boundary terms we
obtain

\begin{align*}
\left(q,-\frac{\partial\rho}{\partial t}-\nabla\cdot\left(\mathbf{\rho u}_{h}\mathbf{+\rho u}_{s}\right)\right) & =\left(q,-\frac{\partial\rho}{\partial t}-\nabla\cdot\left(\mathbf{\rho u}_{h}\right)\right)+\left(\nabla q,\mathbf{\rho u}_{s}\right)
\end{align*}


Putting together all these expressions results in the following equation

\begin{align*}
\left(\mathbf{w},f\right)-\left(\mathbf{w},\frac{\partial\rho(\mathbf{u_{h}+\mathbf{u}_{s}})}{\partial t}\right)\\
-\left(\mathbf{\nabla w},C\nabla^{s}\mathbf{u}_{h}\right)+\int_{\Gamma}\mathbf{w}\cdot2\mu\nabla^{s}\mathbf{u}_{h}\mathbf{\cdot n}+\left(\mathbf{\nabla\cdot w},p_{h}\right)+\left(\mathbf{\nabla\cdot w},p_{s}\right)-\int_{\Gamma}p_{h}\mathbf{w}\mathbf{\cdot n} & +\\
\left(q,-\frac{\partial\rho}{\partial t}-\nabla\cdot\left(\mathbf{\rho u}_{h}\right)\right)+\left(\nabla q,\mathbf{\rho u}_{s}\right) & =0
\end{align*}


Grouping terms containing the effect of the subscales together we
obtain

\begin{align*}
\left(\mathbf{w},f\right)-\left(\mathbf{w},\frac{\partial(\rho\mathbf{u_{h}})}{\partial t}\right)-\left(\mathbf{\nabla w},\mathbf{C}\nabla\mathbf{u}_{h}\right)+\left(\mathbf{\nabla\cdot w},p_{h}\right)+\int_{\Gamma}\mathbf{w}\cdot\mathbf{C}\nabla^{s}\mathbf{u}_{h}\mathbf{\cdot n}-\int_{\Gamma}p_{h}\mathbf{w}\mathbf{\cdot n} & +\\
+\left(\mathbf{w},-\frac{\partial(\mathbf{\mathbf{\rho u}_{s}})}{\partial t}\right)+\left(\mathbf{w},\nabla\cdot\mathbf{C}\nabla^{s}\mathbf{u_{s}}\right)+\left(\mathbf{\nabla\cdot w},p_{s}\right) & +\\
\left(q,-\frac{\partial\rho}{\partial t}-\nabla\cdot\left(\mathbf{\rho u}_{h}\right)\right)+\left(\nabla q,\mathbf{\rho u}_{s}\right) & =0
\end{align*}


We now need to introduce a model for the subscales (**ASGS**)

$\mathbf{u}_{s}=\tau_{1}\left(\mathbf{f}-\rho\frac{\partial\mathbf{u}_{\mathbf{h}}}{\partial t}-\nabla p_{h}\right)$

$p_{s}=\tau_{2}\left(-\frac{\partial\rho}{\partial t}-\nabla\cdot\left(\mathbf{\rho u}_{h}\right)\right)$

We will make the approximation $\frac{\partial\mathbf{u}_{s}}{\partial t}\approx\mathbf{0}$
(quasi-static subscales) and that the fluid is incompressible with
uniform density ($\frac{\partial\rho}{\partial t}=0$ and $\rho=const$
in space)


Let's now focus on the remaining stabilization terms one by one

If we use linear elements, all higher-than-one order derivatives of
finite element functions within each element can be discarded and
we have

$\mathbf{\left(\mathbf{w},\nabla\cdot\mathbf{C}\nabla^{s}\mathbf{u_{s}}\right)}=-\mathbf{\left(\mathbf{\nabla w},\mathbf{C}\nabla^{s}\mathbf{u_{s}}\right)+\int_{\Gamma}...}\approx0.$

## STAB PRESS TERM
$\left(\mathbf{\nabla\cdot w},p_{s}\right)\rightarrow-\tau_{2}\left(\nabla\cdot\mathbf{w},\left(\nabla\cdot\left(\mathbf{\rho u}_{h}\right)\right)\right)$

Pressure Equation Terms

$\left(\nabla q,\rho\mathbf{u_{s}}\right)\rightarrow\left(\nabla q,\rho\tau_{1}\left(\mathbf{f}-\rho\frac{\partial\mathbf{u}_{h}}{\partial t}-\nabla p_{h}\right)\right)$


## SYMBOLS TO BE EMPLOYED
Shape functions $N_{I}$ and derivatives $\nabla N_{I}$stored respectively
in a vector $\mathbf{N}$and a matrix $\mathbf{DN}$

we define the following matrices,

- $\mathbf{P}$ such that $\mathbf{P_{I}}$ is the pressure of node I
- $\mathbf{V}$ such that $\mathbf{V_{Ik}}$ is the velocity of node I,component K
- $\mathbf{W}$ such that $\mathbf{W_{I}}=\mathbf{w_{I}}$
- $\mathbf{Q}$ such that $\mathbf{Q_{I}=}q_{I}$
- $\mathbf{Vn}$ such that $\mathbf{Vn_{Ik}}$ is the veocity of node I, component K (OF THE OLD TIME STEP)
- $\mathbf{Vnn}$ such that $\mathbf{Vnn_{Ik}}$is the veocity of node I, component K (TWO STEPS IN THE PAST)


Values on the gauss points are expressed in terms of such variables
as

- $\mathbf{acc_{h}:=}bdf_{0}\mathbf{V}+bdf_{1}\mathbf{V}_{n}+bdf_{2}\mathbf{V}_{nn}$a
3*1 matrix
- $\mathbf{v_{h}:=\left(V\right)^{T}N}$ a 3*1 matrix
- $\mathbf{w_{h}:=W^{T}N}$ a 3*1 matrix
- $p_{h}\mathbf{:=}\mathbf{P^{T}N}$ a 1*1 matrix
- $\mathbf{q_{h}:=Q^{T}N}$ a 1*1 matrix
- $\mathbf{grad\_ph}:=\mathbf{DN^{T}\cdot P}$ a 3*1 matrix
- $\mathbf{grad\_vh:=DN^{T}\cdot V}$ a 3*3 matrix
- $\mathbf{grad\_qh}:=\mathbf{DN^{T}\cdot Q}$ a 3*1 matrix
- $\mathbf{grad\_wh:=DN^{T}\cdot W}$ a 3*3 matrix
- $\mathbf{\epsilon:=grad\_sym\_vh:=}\sum\mathbf{\mathbf{B_{I}V_{I}}}$
a 3{*}3 matrix (it is the symmetric gradient IN VOIGT FORM!)
- $\mathbf{conv:=\left(v_{h}\cdot grad\_vh\right)^{T}}$ a 3*1 vector

we also introduce the constitutive matrix $\mathbf{C}$(symmetric
6*6 in 3D) and the matrix $\mathbf{B}$ (6*12 for a tetra) such
that $\sigma=\mathbf{B}\epsilon$


## IMPLEMENTATION - GALERKIN PART
The term

$\left(\mathbf{w},f-\frac{\partial\rho u_{h}}{\partial t}+\nabla\cdot\mathbf{C}\nabla^{s}\mathbf{u_{h}}-\nabla p_{h}\right)$

Is implemented on each gauss point as

$\mathbf{w_{h}^{T}}\mathbf{f-}\mathbf{w_{h}^{T}\rho acc_{h}}-\mathbf{grad\_sym\_wh^{T}C}\mathbf{grad\_sym\_vh}+\mathbf{div\_}\mathbf{w_{h}^{T}ph}$


## IMPLEMENTATION-STABILIZATION PART

**PRESSURE TERMS** $-\rho\tau_{2}\mathbf{div\_wh}\mathbf{div\_vh}+\mathbf{grad\_qh^{T}}\mathbf{\rho\tau}_{1}\mathbf{\left(\mathbf{fh-\rho acc_{h}-grad\_ph}\right)}$


## IMPLEMENTATION - ENRICHMENT PART
here we put terms with penr $p_{enr}$ tested against $\mathbf{w}$
and $p_{h}$ and terms in $\mathbf{u_{h},p_{h}}$tested against $q_{enr}$

\begin{align*}
\left(\mathbf{\nabla\cdot w},p_{enr}\right)-\int_{\Gamma}p_{enr}\mathbf{w}\mathbf{\cdot n} & +\\
-\left(\nabla q,\rho\tau_{1}\nabla p_{enr}\right)+\left(\nabla q_{enr},\rho\tau_{1}\left(\mathbf{f}-\rho\frac{\partial\mathbf{u}_{\mathbf{h}}}{\partial t}-\nabla p_{h}-\nabla p_{enr}\right)_{s}\right) & =0
\end{align*}


which is implemented in the practice as

$\mathbf{div\_}\mathbf{w_{h}^{T}penr}+\mathbf{grad\_qh^{t}}\mathbf{\rho\tau}_{1}\mathbf{\left(\mathbf{-grad\_penr}\right)}+\mathbf{grad\_qenr^{T}}\rho\mathbf{\tau}_{1}\mathbf{\left(\mathbf{fh-\rho acc_{h}-grad\_ph-grad\_penr}\right)}$
